```python
def factorial(num):
    """
    this is to find the factorial of a number
    """
    if num ==1:
        return 1
    else:
        return num*factorial(num-1)
factorial(4)
```

```python
def factorial(num):
    """
    this is to find the factorial of a number
    """
    return 1 if num ==1 else return num*factorial(num)

factorial(4)
```

```python
#calculator
import sys
while True:
    total=0
    try:
        num1=int(input("enter the first number"))
        num2=int(input("enter the 2nd number"))
        op=input("enter the operator[+,-,*,/,%]")
        if op=="+":
            total=num1+num2
            print("{}{}{}={}".format(num1,op,num2,total))
            break
        elif op=="-":
            total=num1-num2
            print("{}{}{}={}".format(num1,op,num2,total))
            break
            exit
        elif op=="*":
            total=num1*num2
            print("{}{}{}={}".format(num1,op,num2,total))
            break
        elif op=="%":
            total=num1%num2
            print("{}{}{}={}".format(num1,op,num2,total))
            break
            exit
        elif op=="/":
            try:
                total=num1/num2
                print("{}{}{}={}".format(num1,op,num2,total))
                break
                
            except ZeroDivisionError:
                print("zero cannot be divided")
        else:
            print("invalid operator")
    except:
        print("oops",sys.exc_info(),"occured")
```
